```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model 
from sklearn.model_selection import train_test_split
```

```python
mpg_df=pd.read_csv('Car-mpg- Dataset (1).csv')
```

```python
mpg_df.head(50)
```

```python
mpg_df=mpg_df.drop('car_name',axis=1)
```

```python
mpg_df.head(50)
```

```python
mpg_df['origin']=mpg_df['origin'].replace({1:'america',2:'australia',3:'asia'})
```

```python
mpg_df.head(50)
```

```python
mpg_df=pd.get_dummies(mpg_df,columns=['origin'])
```

```python
mpg_df.head(50)
```

```python
mpg_df.describe().transpose()
```

```python
temp=pd.DataFrame(mpg_df.hp.str.isdigit())
```

```python
temp[temp['hp']==False]
```

```python
mpg_df['hp']=mpg_df['hp'].replace('?',np.nan)
```

```python
mpg_df[mpg_df.isnull().any(axis=1)]
```

```python
mpg_df.median()
```

```python
mpg_df=mpg_df.apply(lambda x:x.fillna(x.median()),axis=0)
```

```python
mpg_df.dtypes
```

```python
mpg_df['hp']=mpg_df['hp'].astype('float64')
```

```python
X=mpg_df.drop('mpg',axis=1)
y=mpg_df[['mpg']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
for idx,col_name in enumerate(X_train.columns):
    print('The coefficients of {} is {}'.format(col_name,regression_model.coef_[0][idx]))
```

```python
intercept=regression_model.intercept_[0]
print('the intercept of our model is {}'.format(intercept))
```

```python
regression_model.score(X_test,y_test)
```
