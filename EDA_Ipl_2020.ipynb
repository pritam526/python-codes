Exploratory Data Analysis of Indian Premier League  **2020**

The 2020 Indian Premier League, also known as IPL 13 and branded as Dream11 Indian Premier League 2020, was the thirteenth season of the IPL, a professional Twenty20 cricket (T20) league established by the Board of Control for Cricket in India (BCCI) in 2007.
The tournament was played between 19 September 2020 to 10 November 2020 in the United Arab Emirates.

We Need to upload the data ,clean data,Import libraries and preprocessing.

```
Ipl_df.columns
```

The dataset has 19 columns
Match_No : Number of Matches

Match : Which team verses which team(Shortform)

Match_FullForm : Which team verses which team(Fullform)

Date : Date of match played

Day : Day of match played

Time_India (IST) : Match Time in Indian standard time

Time_UAE : Match Time in United Arab Emirates

Venue : Venue(Stadium) where matches played

Team1 : One of the current IPL Team

Team2 : One of the current IPL Team

Toss_Winner: Team which won the toss

Toss_Decision : Toss winner's decision to Bat/Field

dl_applied : (1 or 0)indicates whether the Duckworth-Lewis 

rule was applied or not.

winner : The winner of the match.

win_by_wickets : Provides the number of wickets by which the team batting second won.

win_by_runs : Provides the runs by which the team batting first won

player_of_the_match : Game Changing play from player gets this award

Result : The result of the match (‘normal’, ‘tie’)

Super Over : One Over Eliminator,determine the winner of a limited-overs match in case the scores of both teams at the end of the match are tied

## Heat Map : Null values

```
plt.figure(figsize=(12,5))
sns.heatmap(Ipl_df.isnull(),yticklabels=False, annot=True)
plt.xticks(size=15)
plt.show()
```

Observation: In the above heatmap we can see there is no null values in any attribute of the dataset.

## Count plot : Analysing team with the most number of wins in IPL season 13.

```
plt.figure(figsize=(12,7))
ax =sns.countplot('winner',data =Ipl_df,order = Ipl_df['winner'].value_counts().index,palette='viridis')
plt.title("Total number of wins by each team till 2020",size=18)
plt.xticks(rotation=45, ha = 'right',size=13)
plt.yticks(np.arange(13, step=2),size=13)
plt.xlabel('Teams',size=15)
plt.ylabel('Number of matches',size=15)

def annotation_plot(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.1f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+0.2))
annotation_plot(ax,0.08,1)
plt.show()
```

As we can observe from the Above graph that which team has won most number of matches in ipl 2020

# Which Team Had won the most number of Tosses.

```
team_wins = Ipl_df['Toss_Winner'].value_counts()

team_wins_df = pd.DataFrame(columns=["team", "tosswins"])
for items in team_wins.iteritems():
    temp_df1 = pd.DataFrame({
        'team':[items[0]],
        'tosswins':[items[1]]
    })
    team_wins_df = team_wins_df.append(temp_df1, ignore_index=True)
```

```
team_wins_df
```

```
plt.figure(figsize=(12,8))
ax=sns.barplot(x='team', y='tosswins', data=team_wins_df, palette='inferno',saturation=90.75)
plt.title('Team wise total toss wins',size =20)
plt.xlabel('Teams',size=15)
plt.ylabel('Total No of toss wins',size=15)
plt.yticks(np.arange(13,step=1),size=15)
plt.xticks(size=13)
def annotation_plot(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.0f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+0.2))
annotation_plot(ax,0.19,1)

plt.show()
```

Observation : Toss plays vital role in any matches considering dew factor, By above bar graph, it's clear that SRH team has been lucky this season, as they are highest toss winner and KXIP are least times toss winner.

## **Who won most of 'player of match 'in the tournament.**

```
Ipl_df['player_of_the_match'].value_counts(ascending=True)
```

```
mvp_ser = Ipl_df['player_of_the_match'].value_counts()

mvp_ten_df = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
    temp_df2 = pd.DataFrame({
            'player':[items[0]],
            'wins':[items[1]]
        })
    mvp_ten_df = mvp_ten_df.append(temp_df2, ignore_index=True)
    count += 1    
```

```
plt.figure(figsize=(12,12))
plt.title("Players who won 'Player of the Match'",size=20)
ax=sns.barplot(x='wins', y='player', data=mvp_ten_df, palette='magma',saturation=20.75);
plt.xticks(np.arange(4, step=1),size=15)
plt.xlabel("No of Man of the match",size=15)
plt.yticks(size=13)
plt.ylabel("Players",size=15)

plt.show()
```

Observation : In the above graph x axis presenting number of times the player has won player of match award and y axis presenting player names, in this analysis players named KL Rahul, Ruturaj Gaikwad and AB de villiers has maximum number of player of match award( ie 4 times) and followed by SuryaKumar Yadav, Shikar Dhawan, Ben Stokes, Anrich Nortje,Quinton de kock, Marcus stonis, Yuzvendra Chahal and Sanju Samson has won second highest times(ie 3 times) and rest players has got atmost 1 time.

```
mvp_ser = Ipl_df['player_of_the_match'].value_counts()

mvp_ten_df = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
  if count>9:
    break
  else:

    temp_df2 = pd.DataFrame({
            'player':[items[0]],
            'wins':[items[1]]
        })
    mvp_ten_df = mvp_ten_df.append(temp_df2, ignore_index=True)
    count += 1  
```

```
plt.figure(figsize=(12,7))
plt.title("Top 10 Best Performer in 2020 IPL",size=20)
sns.barplot(x='wins', y='player', data=mvp_ten_df, palette='magma',saturation=20.75);
plt.xticks(np.arange(4, step=1),size=13)
plt.yticks(size=13)
plt.xlabel("No of Man of the match",size=1)

plt.ylabel("Players",size=15)

plt.grid()
plt.show()
```

## Which were the Top 10 Rivalaries Head to Head.

```
h2h_counts= Ipl_df.groupby('Match').apply(lambda x:x['Match'].count()).reset_index(name='Match Counts')
top_h2h_order=h2h_counts.sort_values(by='Match Counts',ascending=False)
top_h2h=top_h2h_order[:10]
#print('Head to Head Rivalry\n',top_h2h)
plt.figure(figsize=(7,7))
plt.pie(top_h2h['Match Counts'],labels=top_h2h['Match'],autopct='%1.1f%%', startangle=30)
plt.axis('on')
plt.title('Top 10 most head to head Rivalry Matches Played',size=20)

plt.show()
```

## **Scatter Plot :Match Result Analysis**

```
a_list = Ipl_df['win_by_runs'].tolist()#win_by_runs
a_list=[x + 80 for x in a_list]
plt.figure(figsize=(12,10))
ax = sns.scatterplot(x="Match_No",y='Match',s=a_list,data = Ipl_df,hue="winner",palette='viridis')
ax.text(60.7,50, "IPL 2020 Final Match(MI vs DC)", horizontalalignment='left', size='medium', color='black', weight='semibold')
ax.text(56.7,52, "Eliminator Match(RCB vs SRH)", horizontalalignment='left', size='medium', color='black', weight='semibold')
ax.text(50,27, "Qualifer 1 Match(DC vs MI)", horizontalalignment='left', size='medium', color='black', weight='semibold')
ax.text(48,11, "Qualifer 2 Match(DC vs SRH)", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.title("Match wise analysis of match results",size=18)
plt.xticks(np.arange(60, step=2),size=15)
plt.yticks(size=10)
plt.xlabel('Match No',size=15)
plt.ylabel('Matchs',size=15)
ax.legend(loc="center", bbox_to_anchor=(1.08, 0.33), ncol=1)
plt.show()
```

## **Heat Map:Team wise win count at the Venue**

```
venue_win=Ipl_df.groupby(['Venue','winner']).apply(lambda x:x['winner'].count()).reset_index(name='Match Wins')
venue_win_pvt=pd.pivot(venue_win,values='Match Wins',index='Venue',columns='winner')
venue_win_pvt.replace(np.NaN,0)
plt.figure(figsize=(20,6))
htmp=sns.heatmap(venue_win_pvt,annot=True,fmt='g',cmap='inferno')
plt.xticks(size=13)
plt.yticks(size=13)
plt.xlabel('Teams',size=17)
plt.ylabel('Venues',size=17)
plt.title('Team wise win count at the Venues',size=20)

plt.show()
```
